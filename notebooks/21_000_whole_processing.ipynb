```python
import re
import nltk

from src.file import read

nltk.download('stopwords')
from nltk.corpus import stopwords

nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize

nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
```

### Opening file

```python
content = read("frank_2nd_para")
print(content)
```

### Lowercasing

```python
lower_content = ' '.join([txt.lower() for txt in content.split()])
print(lower_content)
```

### Removing stopwords

```python
en_stopwords = stopwords.words('english')
print(en_stopwords)
```

```python
content_no_stopwords = ' '.join([word for word in lower_content.split() if word not in en_stopwords])
print(content_no_stopwords)
```

### Removing punctation

```python
pattern_to_find = r"[^\w\s]"
# [^ ] means "not", \w means word and \s means whitespace: so find anything that is not a word or a space

content_no_punctation = re.sub(pattern_to_find, "", content_no_stopwords)
print(content_no_punctation)
```

### Tokenization

```python
tokens = word_tokenize(content_no_punctation)
print(tokens)
```

### Lemmatization

```python
lem = WordNetLemmatizer()
content_lemmatized = [lem.lemmatize(token) for token in tokens]
print(content_lemmatized)
```
