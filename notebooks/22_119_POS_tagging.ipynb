```python
import re
import spacy
import pandas as pd

from src.file import read
```

```python
content = read("frank_2nd_para")
print(content)
```

```python
lower_content = ' '.join([txt.lower() for txt in content.split()])
print(lower_content)
```

```python
pattern_to_find = r"[^\w\s]"
content_no_punctation = re.sub(pattern_to_find, "", lower_content)
print(content_no_punctation)
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
spacy_doc = nlp(content_no_punctation)
```

```python
pos_df = pd.DataFrame(columns=['token', 'pos_tag'])
```

```python
for token in spacy_doc:
    pos_df = pd.concat(
        [
            pos_df,
            pd.DataFrame.from_records([{
                'token': token.text,
                'pos_tag': token.pos_}])
        ],
        ignore_index=True)
pos_df.head()
```

```python
pos_df_counts = pos_df \
    .groupby(['token', 'pos_tag']) \
    .size() \
    .reset_index(name='counts') \
    .sort_values(by='counts', ascending=False)
pos_df_counts.head(10)
```

```python
pos_df_pos_counts = pos_df_counts \
    .groupby(['pos_tag'])['token'] \
    .count() \
    .sort_values(ascending=False)
pos_df_pos_counts.head(10)
```

```python
nouns = pos_df_counts[pos_df_counts.pos_tag == 'NOUN'][:10]
nouns
```
