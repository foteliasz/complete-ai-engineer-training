```python
import re
import nltk

nltk.download('stopwords')
from nltk.corpus import stopwords

nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize

from nltk.stem import PorterStemmer
```

### Opening file

```python
with open('frank_2nd_para', 'r') as content_file:
    content = content_file.read()

print(content)
```

### Lowercasing

```python
lower_content = ' '.join([txt.lower() for txt in content.split()])
print(lower_content)
```

### Removing stopwords

```python
en_stopwords = stopwords.words('english')
print(en_stopwords)
```

```python
content_no_stopwords = ' '.join([word for word in lower_content.split() if word not in en_stopwords])
print(content_no_stopwords)
```

### Removing punctation

```python
pattern_to_find = r"[^\w\s]"
# [^ ] means "not", \w means word and \s means whitespace: so find anything that is not a word or a space

content_no_punctation = re.sub(pattern_to_find, "", content_no_stopwords)
print(content_no_punctation)
```

### Tokenization

```python
tokens = word_tokenize(content_no_punctation)
print(tokens)
```

### Stemming

```python
ps = PorterStemmer()
content_stemmed = [ps.stem(token) for token in tokens]
print(content_stemmed)
```
