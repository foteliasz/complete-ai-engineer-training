```python
# import packages
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
# assign our stop words to a variable
en_stopwords = stopwords.words('english')

# print the list of stop words to see what we will be removing
print(en_stopwords)
```

```python
with open('frank_1st_para', 'r') as content_file:
    content = content_file.read()

print(content)
```

```python
# keep the words in the sentence if the word is not in the list of stop words
content_no_stopwords = ' '.join([word for word in content.split() if word not in en_stopwords])
print(content_no_stopwords)
```
